```python
# Start your codes here!
# import some libs

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

# read file csv
terr = pd.read_csv('terrorism.csv', encoding='ISO-8859-1')
terr.head(5)
```

```python
terr.describe()
```

```python
terr.isnull().sum()
```

```python

terr_iyear = terr.iyear.value_counts().sort_index()

terr_imonth = terr.imonth.value_counts().sort_index()
imonth_index = np.array(terr_imonth.index)
imonth_values = np.array(terr_imonth.values)

fig, ax = plt.subplots(2, figsize=(12,9))
ax[0].plot(terr_iyear)
ax[0].set_title('Terrorism by Year')
ax[1].bar(imonth_index, imonth_values)
ax[1].set_title('Terrorism by Month')
fig.tight_layout(pad=2.0)
plt.show()
```

```python
import seaborn as sns
plt.subplots(figsize=(18,9))
sns.countplot('attacktype1_txt', data=terr, order=terr['attacktype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Attack Method')
plt.show()

plt.subplots(figsize=(18,9))
sns.countplot('targtype1_txt', data=terr, order=terr['targtype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Main Targets')
plt.show()

plt.subplots(figsize=(18,9))
sns.countplot('region_txt', data=terr, order=terr['region_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Terrorism by Region')
plt.show()

```

```python
print("The highest terrorist attacks in ", terr.country_txt.value_counts().index[0])
```

```python
coun_terr=terr['country_txt'].value_counts()[:15].to_frame()
coun_terr.columns=['attacktype1_txt']
coun_kill=terr.groupby('country_txt')['nkill'].sum().to_frame()

terr.effected = terr.nkill.fillna(0) + terr.nwound.fillna(0)
coun_effected = terr.groupby('country_txt')['effected'].sum().to_frame()

coun_terr.merge(coun_kill,left_index=True,right_index=True,how='left').merge(coun_effected,left_index=True,right_index=True,how='left').plot.bar(width=0.9)
fig=plt.gcf()
fig.set_size_inches(25,6)
plt.show()
```

```python
top_groups10=terr[terr['gname'].isin(terr['gname'].value_counts()[1:11].index)]
pd.crosstab(top_groups10.iyear,top_groups10.gname).plot()
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.title('Activities of Top Terrorist Groups')
plt.show()
```

```python
top_type=terr[terr['attacktype1_txt'].isin(terr['attacktype1_txt'].value_counts().index)]
pd.crosstab(top_type.iyear,top_type.attacktype1_txt).plot()
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.title('Favourite Attacks Type')
plt.show()
```
